```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification

# Create a small synthetic dataset
X, y = make_classification(
    n_samples=100, 
    n_features=5, 
    n_classes=2, 
    weights=[0.9, 0.1],  # 90% in class 0, 10% in class 1
    random_state=42
)

# Convert the array to a DataFrame
df = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(1, 6)])
df['Stress Level'] = y

# Display the original dataset
print("Original Dataset:")
print(df['Stress Level'].value_counts())
print(df.head())

# Apply SMOTE for oversampling
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new DataFrame with the resampled data
resampled_df = pd.DataFrame(X_resampled, columns=[f'Feature_{i}' for i in range(1, 6)])
resampled_df['Stress Level'] = y_resampled

# Display the resampled dataset
print("\nResampled Dataset:")
print(resampled_df['Stress Level'].value_counts())
print(resampled_df.head())
```
